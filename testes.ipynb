```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df=pd.read_csv("epa-sea-level.csv")
df
```

```python
plt.plot(df["Year"],df["CSIRO Adjusted Sea Level"], "o")
```

```python
first=linregress(df["Year"],df["CSIRO Adjusted Sea Level"])
plt.plot(df["Year"],df["CSIRO Adjusted Sea Level"], "o")
plt.plot(df["Year"], first.intercept + first.slope*df["Year"], 'r', label='fitted line')
```
